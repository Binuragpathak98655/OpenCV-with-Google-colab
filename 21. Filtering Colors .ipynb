```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Define our imshow function 
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()

# Download and unzip our images and Haarcascade Classifiers
!wget https://moderncomputervision.s3.eu-west-2.amazonaws.com/images.zip

!unzip -qq images.zip
```

```python
image = cv2.imread('/content/images/truck.jpg')

lower = np.array([90,0,0])
upper = np.array([135,255,255])

hsv_img = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

mask = cv2.inRange(hsv_img, lower, upper)

res = cv2.bitwise_and(image, image, mask=mask)

imshow('mask',mask)
imshow('hsv imagw', hsv_img)
imshow('Origianal', image)
imshow('Filtered Color only', res)
```

```python
image = cv2.imread('/content/images/Hillary.jpg')


img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# lower mask (0-10)
lower = np.array([0,0,0])
upper = np.array([10,255,255])
mask0 = cv2.inRange(img_hsv, lower, upper)

#upper mask (170-180)
lower_red = np.array([170,0,0])
upper_red = np.array([180,255,255])
mask1 = cv2.inRange(img_hsv, lower_red, upper_red)
#joim mask
mask = mask0 + mask1

res = cv2.bitwise_and(image, image, mask=mask)

imshow('Filter color ', res)
imshow('mask', mask)
imshow('Original', image)
```
